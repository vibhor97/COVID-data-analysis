```python
import pandas as pd
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df.head()
```

```python
df = df[["sno","state_name","active","positive","cured","death","new_active","new_positive","new_cured","new_death","state_code"]]
```

```python
df
```

```python
df = df.iloc[:-1,1:]
```

```python
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
total_states = np.arange(len(df['state_name']))
```

```python
total_states
```

```python
df['active']
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(7, 10), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['positive'], align='center', alpha=0.5,  
                 color=(1,0.5,0),  
                 edgecolor=(1,0.9,1))
    
plt.yticks(total_states, df['state_name'])  
plt.xlim(1,max(df['positive'])+100000) 
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(7, 7.5), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['positive'], align='center', alpha=0.5,  
                 color=(1,0,0),  
                 edgecolor=(1,0.9,1))
    
plt.yticks(total_states, df['state_name'])  
plt.xlim(1,max(df['active'])+1000000) 
plt.xlabel('Active Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```
